```python
import numpy as np
from os import listdir
import operator
```

将文件信息读取为向量 
32\*32 ——> 1\*1024

```python
def img2vector(filename):
    # 创建一个 1*1024 的的数组 
    returnVect = np.zeros((1,1024))
    fr = open(filename)
    for i in range(32):
        # 读取文件
        lineStr = fr.readline()
        for j in range(32):
            # 存储每行前32个字符值
            returnVect[0,32*i+j] = int(lineStr[j])
    return returnVect
```

k-近邻算法

```python
def classify0(inX, dataSet, labels, k):
    dataSetSize = dataSet.shape[0]
    # 计算欧式距离
    diffMat = np.tile(inX, (dataSetSize, 1)) - dataSet
    sqDiffMat = diffMat**2
    distances = np.sqrt(sqDiffMat.sum(axis=1))
    
    sortedDistIndicies = distances.argsort()
    classCount = {}
    # 选择距离最小的 k 个点
    for i in range(k):
        voteIlabel = labels[sortedDistIndicies[i]]
        classCount[voteIlabel] = classCount.get(voteIlabel, 0) + 1
    # 排序
    sortedClassCount = sorted(classCount.items(), key=operator.itemgetter(1), reverse=True)
    return sortedClassCount[0][0]
```

```python
def handwritingClassTest():
    hwLabels = []
    # 读取样本文件列表
    trainingFileList = listdir('F:/workspace-Jupyter/2_kNN/digits/trainingDigits')
    m = len(trainingFileList)
    trainingMat = np.zeros((m,1024))    # 创建m行1024列训练矩阵
    for i in range(m):
        # 根据文件名获得样本标签
        fileNameStr = trainingFileList[i]
        fileStr = fileNameStr.split('.')[0]
        classNumStr = int(fileStr.split('_')[0])
        hwLabels.append(classNumStr)
        # 图像转化为向量
        trainingMat[i, :] = img2vector('F:/workspace-Jupyter/2_kNN/digits/trainingDigits/%s' % fileNameStr)
    # 读取验证文件列表
    testFileList = listdir('F:/workspace-Jupyter/2_kNN/digits/testDigits')
    errorCount = 0.0
    mTest = len(testFileList)
    for i in range(mTest):
        # 根据文件名获得样本标签
        fileNameStr = testFileList[i]
        fileStr = fileNameStr.split('.')[0]
        classNumStr = int(fileStr.split('_')[0])
        # 图像转化为向量
        vectorUnderTest = img2vector('F:/workspace-Jupyter/2_kNN/digits/testDigits/%s' % fileNameStr)
        # kNN分类
        classifierResult = classify0(vectorUnderTest, trainingMat, hwLabels, 3)
        print("the class came back with: %d, the real answer is: %d" % (classifierResult, classNumStr))
        if(classifierResult != classNumStr):
            errorCount += 1.0
    print("\nthe total number of errors is: %d" % errorCount)
    print("\nthe total error rate is: %f" % (errorCount/float(mTest)))        
```

```python
handwritingClassTest()
```
